## Line Plot

### Univariate
Visual representation of the relationship of a single column in a dataset.

````python
plt.plot(cars_dataframe["Min_coverage"], cars_dataframe['Min_coverage'])
plt.title("Min Vs. Max")
plt.xlabel("Min. Coverage")
plt.ylabel("Max. Coverage")
# Save chart as image
plt.savefig('Images-Data-Visualization/linePlot_Cars.png', dpi=300, bbox_inches='tight')
plt.show()
````

## Scatter Plot

### Bivariate
Visual representation of the relationship between two variables in a dataset.

````python
plt.grid(True)
plt.scatter(cars_dataframe["Cost_of_vehicle"], cars_dataframe["Max_coverage"])
plt.title("Cost vs Max coverage")
plt.xlabel("Cost ($)")
plt.ylabel("Max Coverage")
# Save chart as image
plt.savefig('Images-Data-Visualization/scatterPlot_Cars.png', dpi=300, bbox_inches='tight')
````

### Multivariate
Visual representation of the relationship between three or more columns/ variables in a dataset.

## Histograms

````python
sns.histplot(data=cars_dataframe, x="Cost_of_vehicle")
plt.title("Overall Analysis of Vehicles")
plt.xlabel("Cost of Vehicle")
plt.ylabel("Maximum Coverage")

# Save chart as image
plt.savefig('Images-Data-Visualization/histogram_Cars.png', dpi=300, bbox_inches='tight')
````

````python
# Keep only numeric columns
numeric_df = cars_dataframe.select_dtypes(include='number')

# Now compute the correlation
correlation = numeric_df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')


# Save chart as image
plt.savefig('Images-Data-Visualization/corelation_heatMap_Cars.png', dpi=300, bbox_inches='tight')

plt.show()
````
